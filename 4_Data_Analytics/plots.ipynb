```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(
    "/Users/rmn/repos/ibm_professional_data_engineering_certificate_final_capstone_project/4_Data_Analytics/ecommerce.csv"
)
df.head(2)
```

```python
df_2020 = df.loc[df.year == 2020]
monthly_total_price = df_2020.groupby("month")["price"].sum().reset_index()
monthly_total_price

# Create a bar plot
plt.plot(monthly_total_price["month"], monthly_total_price["price"])

# Add labels and title
plt.xlabel("Month")
plt.ylabel("Total Price")
plt.title("Total Price per Month in 2020")

# Display the plot
plt.show()
```

```python
df_pie = df.groupby("category")["price"].sum().reset_index()

plt.pie(df_pie.price, labels=df_pie.category)
plt.title("Total Price by Category")

plt.show()
```

```python
df_mobiles = df.loc[df.item == "Mobile"]
df_agg = df_mobiles.groupby(by=["year", "quarter"])["price"].sum().reset_index()

plt.bar(x=df_agg.quarter, height=df_agg.price)
plt.xticks([1, 2, 3, 4])
plt.xlabel("Quarter")
plt.title("Mobile Sales by Quarter")
plt.show()
```
